```python
import os
import shutil
import pandas as pd

# Load the dataset, skipping bad lines
styles_df = pd.read_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/fashion-dataset/styles.csv', on_bad_lines='skip')

# Filter the data for tops and bottoms
male_tops_df = styles_df[(styles_df['gender'] == 'Men') & (styles_df['subCategory'] == 'Topwear')]
female_tops_df = styles_df[(styles_df['gender'] == 'Women') & (styles_df['subCategory'] == 'Topwear')]
male_bottoms_df = styles_df[(styles_df['gender'] == 'Men') & (styles_df['subCategory'] == 'Bottomwear')]
female_bottoms_df = styles_df[(styles_df['gender'] == 'Women') & (styles_df['subCategory'] == 'Bottomwear')]

# Define the paths
images_folder = 'C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/fashion-dataset/images'
output_folder = 'C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/filtered_images'

# Create directories for filtered images
os.makedirs(output_folder, exist_ok=True)
os.makedirs(os.path.join(output_folder, 'male_tops', 'formal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'male_tops', 'informal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'female_tops', 'formal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'female_tops', 'informal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'male_bottoms', 'formal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'male_bottoms', 'informal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'female_bottoms', 'formal'), exist_ok=True)
os.makedirs(os.path.join(output_folder, 'female_bottoms', 'informal'), exist_ok=True)

def copy_images(df, category, subcategory):
    for idx, row in df.iterrows():
        img_filename = f"{row['id']}.jpg"
        src_path = os.path.join(images_folder, img_filename)
        dest_path = os.path.join(output_folder, category, subcategory, img_filename)
        
        if os.path.exists(src_path):
            shutil.copy(src_path, dest_path)

# Filter for formal and informal outfits and copy images
formal_male_tops_df = male_tops_df[male_tops_df['usage'] == 'Formal']
informal_male_tops_df = male_tops_df[male_tops_df['usage'] != 'Formal']
copy_images(formal_male_tops_df, 'male_tops', 'formal')
copy_images(informal_male_tops_df, 'male_tops', 'informal')

formal_female_tops_df = female_tops_df[female_tops_df['usage'] == 'Formal']
informal_female_tops_df = female_tops_df[female_tops_df['usage'] != 'Formal']
copy_images(formal_female_tops_df, 'female_tops', 'formal')
copy_images(informal_female_tops_df, 'female_tops', 'informal')

formal_male_bottoms_df = male_bottoms_df[male_bottoms_df['usage'] == 'Formal']
informal_male_bottoms_df = male_bottoms_df[male_bottoms_df['usage'] != 'Formal']
copy_images(formal_male_bottoms_df, 'male_bottoms', 'formal')
copy_images(informal_male_bottoms_df, 'male_bottoms', 'informal')

formal_female_bottoms_df = female_bottoms_df[female_bottoms_df['usage'] == 'Formal']
informal_female_bottoms_df = female_bottoms_df[female_bottoms_df['usage'] != 'Formal']
copy_images(formal_female_bottoms_df, 'female_bottoms', 'formal')
copy_images(informal_female_bottoms_df, 'female_bottoms', 'informal')

# Save the filtered data to new CSV files for reference
formal_male_tops_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/formal_male_tops_outfits.csv', index=False)
informal_male_tops_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/informal_male_tops_outfits.csv', index=False)
formal_female_tops_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/formal_female_tops_outfits.csv', index=False)
informal_female_tops_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/informal_female_tops_outfits.csv', index=False)
formal_male_bottoms_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/formal_male_bottoms_outfits.csv', index=False)
informal_male_bottoms_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/informal_male_bottoms_outfits.csv', index=False)
formal_female_bottoms_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/formal_female_bottoms_outfits.csv', index=False)
informal_female_bottoms_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/informal_female_bottoms_outfits.csv', index=False)
```

```python
import os
import shutil
import pandas as pd

# Load the dataset, skipping bad lines
styles_df = pd.read_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/fashion-dataset/styles.csv', on_bad_lines='skip')

# Filter the data for formal shirts (assuming 'Shirt' is the value in 'articleType')
formal_shirts_df = styles_df[(styles_df['articleType'].str.contains('Shirt', case=False, na=False)) & (styles_df['usage'] == 'Formal')]

# Define the paths
images_folder = 'C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/fashion-dataset/images'
output_folder = 'C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/filtered_images/male_formal_tops'

# Create the output directory for formal shirt images
os.makedirs(output_folder, exist_ok=True)

def copy_images(df):
    for idx, row in df.iterrows():
        img_filename = f"{row['id']}.jpg"
        src_path = os.path.join(images_folder, img_filename)
        dest_path = os.path.join(output_folder, img_filename)
        
        if os.path.exists(src_path):
            shutil.copy(src_path, dest_path)
            print(f"Copied {src_path} to {dest_path}")
        else:
            print(f"Image {src_path} not found")

# Copy the filtered images to the output folder
copy_images(formal_shirts_df)

# Save the filtered data to a new CSV file for reference
formal_shirts_df.to_csv('C:/Users/ARA/Desktop/Outfit-Aura/OUTFIT-AURA-web-application-/backend/data/filtered_images/male_formal_tops/formal_tops_outfits.csv', index=False)
```
